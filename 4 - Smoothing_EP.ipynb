# Smoothing Hyperspectral Data

```python
import hylite
import numpy as np
%matplotlib inline
```

```python
from hylite import io
```

```python
#Loading the data
EnMAP = io.load('/Users/rachdi18/Documents/Namibia_Epembe/Enmap/enmapSUBCUT.hdr' )
PRISMA= io.load('/Users/rachdi18/Documents/Namibia_Epembe/PRISMA/prismaSubset.hdr')
```

```python
# create a copy of the data to avoid overwriting the original image
copy= EnMAP.copy()
```

```python
#Data normalization
EnMAP.data = EnMAP.data.astype(np.float32) / np.nanmax(EnMAP.data)
```

```python
# spectral plot before smoothing
fig,ax = EnMAP.plot_spectra(band_range=(500., 2450.)) # plot it
ax.set_xlim(500., 2450.)
ax.set_xticks(np.arange(500., 2450.,100))
ax.set_xlabel("Wavelength (nm)")
```

```python
#Applies median filter on data
copy.smooth_median(3)
smooth_mf= copy
```

```python
# spectral plot after smoothing
fig,ax = copy.plot_spectra(band_range=(500., 2450.)) # plot it
ax.set_xlim(500., 2450.)
ax.set_xticks(np.arange(500., 2450.,100))
ax.set_xlabel("Wavelength (nm)")
```

```python
#saving the smoothed image
io.save('/Users/rachdi18/Documents/Smoothing/enmap_savgol.hdr', copy)
```
